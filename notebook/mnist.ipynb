## creating MNIST model for identifying clothes

```python
#impoting libraries

import tensorflow as tf


data = tf.keras.datasets.fashion_mnist
```

```python
#loading data

(train_img,train_lbl),(test_img,test_lbl) = data.load_data()
```

```python
#normalizing images (converting imges into gray scale

train_img = train_img/255.0
test_img = test_img/255.0
```

```python
#defining neural network

model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dense(64, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
#compiling a model

model.compile(optimizer='adam',
              loss= 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
#training data

model.fit(train_img,train_lbl, epochs=5)
```

```python
#evaluating data

model.evaluate(test_img,test_lbl)
```
